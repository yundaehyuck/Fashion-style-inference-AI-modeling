```python
import json
import urllib
from PIL import Image
import matplotlib.pyplot as plt
from tqdm.auto import tqdm
```

```python
with open("/content/drive/MyDrive/singledata/상의/vest.json", 'r') as file:
    data = json.load(file)
```

```python
data[0]
```

```python
for i,d in enumerate(tqdm(data),start=4400):
    
    url = d['imageUrl']

    img = Image.open(urllib.request.urlretrieve(url)[0]).convert("RGB")

    img.save(f"/content/drive/MyDrive/single_label_data/data/상의/{i}.jpg")
```
